## Упражнение 2. Сумма первых n положительных чисел

**Условие.** Напишите программу, запрашивающую у пользователя число и подсчитывающую сумму натуральных положительных чисел от 1 до введенного
пользователем значения. Сумма первых n положительных чисел может
быть рассчитана по формуле:

\begin{align}
sum = \frac{(n)(n+1)}{2}

**Перепишите код решения и запустите его**

image.png

```python
# Ваш код

n = int(input('Введите положительное число: '))
nsum = n * (n + 1) / 2
print(('Сумма первых %d положительных чисел равна %d') % (n, nsum))
```

## Упражнение 3. Площадь правильного многоугольника

**Условие.** Многоугольник называется правильным, если все его стороны и углы равны. Площадь такой фигуры можно вычислить по следующей формуле,
в которой s – длина стороны, а n – количество сторон:

\begin{align}
   \frac{n \cdot \mathbf{s^2}}{4 \cdot \mathrm{tan(\frac{\pi}{n})}}
\end{align}

```python
# Вычеслить площадь правильного многоугольника

# Запрашивем информацию у пользователя
s = int(input('Введите длина многоугольника: '))
n = int(input('Введите количество сторон многоугольника: '))

# Вычесляем площадь
import math
t = math.tan
p = math.pi
ns = n * s**2 / 4 * t(p/n)

# Отобразим результат
print(ns)
```
