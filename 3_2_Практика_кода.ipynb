## Упражнение 1. Билеты в зоопарк

**Условие.** В зоопарке цена входного билета зависит от возраста посетителя. Дети до двух лет допускаются бесплатно. Дети в возрасте от трех до 12 лет могут посещать зоопарк за 140. Пенсионерам старше 65 лет вход обойдется в 180, а обычный взрослый билет стоит 230.
Напишите программу, которая будет запрашивать возраст всех посетителей в группе (по одному за раз) и выводить общую цену билетов для посещения зоопарка этой группой. В качестве индикатора окончания ввода можно по традиции использовать пустую строку. Общую цену билетов стоит выводить в формате с двумя знаками после запятой.

**Перепишите код решения и запустите его**

image.png

image.png

```python
# Ваш код

BABY_PRICE = 0.00
CHILD_PEICE = 140
ADULT_PRICE = 230
SENIOR_PRICE = 180

BABY_LIMIT = 2
CHILD_LIMIT = 12
ADULT_LIMIT = 64

total = 0

age_line = input("Введите возраст посетителя (или пустая строка для оканчания ввода): ")
while age_line != "":
    age = int(age_line)

    if age <= BABY_LIMIT:
      total = total + BABY_PRICE
    elif age <= CHILD_LIMIT:
      total = total + CHILD_PEICE
    elif age <= ADULT_LIMIT:
      total = total + ADULT_PRICE
    else:
      total = total + SENIOR_PRICE

    age_line = input("Введите возраст посетителя (или пустая строка для оканчания ввода): ")

print("The total price of visitors: ", total, "₽")
```

## Упражнение 2. Порядок сортировки


**Условие.** Напишите программу, которая будет запрашивать у пользователя целочисленные значения и сохранять их в виде списка. Индикатором окончания ввода значений должен служить ноль. Затем программа должна вывести на экран все введенные пользователем числа (кроме нуля) в порядке возрастания – по одному значению в строке. Используйте для сортировки  метод sort.

**Перепишите код решения и запустите его**

image.png

```python
# Ваш код

num_list = []

num = int(input("Введите целое число (0 для окончания ввода)"))

while num != 0:
    num_list.append(num)
    num = int(input("Введите целое число (0 для окончания вв1ода)"))

num_list.sort()

for num in num_list:
    print(num)
```

```python
# Ваш код

# Собираем уникальные случайные номера для лотерейного билета

from random import randrange

MIN_NUM = 1
MAX_NUM = 49
TOTAL_NUMS = 6

ticket_nums = []

for x in range(TOTAL_NUMS):
    x = randrange(MIN_NUM, MAX_NUM)

    while x in ticket_nums:
        x = randrange(MIN_NUM, MAX_NUM)
    ticket_nums.append(x)


ticket_nums.sort()

# print(*ticket_nums)

for num in ticket_nums:
    print(num, end=" ")
```

```python
# Ваш код

#@title Решение (более элегантное с применением множеств)
##
# Собираем уникальные случайные номера для лотерейного билета
#
from random import randrange
MIN_NUM = 1
MAX_NUM = 49
NUM_NUMS = 6

# Используем список для хранения номеров лотерейного билета
ticket_nums = set()
# Генерируем NUM_NUMS случайных, но уникальных значений
while len(ticket_nums) < NUM_NUMS:
    # Генерируем номер, которого еще нет в списке
    ticket_nums.add(randrange(MIN_NUM, MAX_NUM + 1))

# Сортируем номера по возрастанию и отображаем их
ticket_nums = sorted(ticket_nums)

print("Номера вашего билета: ", end="")
for n in ticket_nums:
    print(n, end=" ")
```

## Упражнение 4. Уникальные символы

**Условие.** Напишите программу, определяющую и выводящую на экран количество
уникальных символов во введенной пользователем строке. Например, в строке Hello, World! содержится десять уникальных символов, а в строке zzz – один. Используйте словарь для решения этой задачи.

**Перепишите код решения и запустите его**

image.png

```python
# Ваш код

s = input("Inter a string: ")

dic_s = {}

for i in s:
  dic_s[i] = True

print("Unique letters are ", len(dic_s))
```
