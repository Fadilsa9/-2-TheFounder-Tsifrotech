```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. تحميل وقراءة البيانات
df = pd.read_csv("/content/drive/MyDrive/🟢TheFounder/vgsale_1.csv")

# 2. تنظيف البيانات
df = df.dropna(subset=['Year', 'Genre', 'Global_Sales'])  # نحذف الصفوف التي ينقصها عام أو نوع
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df = df.dropna(subset=['Year'])  # نحذف صفوف بها أعوام غير قابلة للتحويل
df['Year'] = df['Year'].astype(int)

# 3. فصل البيانات إلى فترتين
df_before_2000 = df[df['Year'] < 2000]
df_after_2000 = df[df['Year'] >= 2000]

# 4. تحليل عدد الألعاب المنشورة حسب النوع
genre_count_before = df_before_2000['Genre'].value_counts()
genre_count_after = df_after_2000['Genre'].value_counts()

# 5. تحليل إجمالي المبيعات حسب النوع
genre_sales_before = df_before_2000.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
genre_sales_after = df_after_2000.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

# print("📊 Number of games by genre before 2000:")
# print(genre_count_before)
# print("\n📊 Number of games by genre after 2000:")
# print(genre_count_after)

# print("\n💰 Total sales by genre before 2000 (in millions):")
# print(genre_sales_before)
# print("\n💰 Total sales by genre after 2000 (in millions):")
# print(genre_sales_after)

#----------------------
### الرسم ###
# 6. رسم النتائج
plt.figure(figsize=(15, 6))

# عدد الألعاب قبل 2000
plt.subplot(2, 2, 1)
sns.barplot(x=genre_count_before.index, y=genre_count_before.values, palette="Blues_d")
plt.title("عدد الألعاب المنشورة حسب النوع قبل 2000")
plt.xticks(rotation=45)
plt.ylabel("عدد الألعاب")

# عدد الألعاب بعد 2000
plt.subplot(2, 2, 2)
sns.barplot(x=genre_count_after.index, y=genre_count_after.values, palette="Greens_d")
plt.title("عدد الألعاب المنشورة حسب النوع بعد 2000")
plt.xticks(rotation=45)
plt.ylabel("عدد الألعاب")

# المبيعات قبل 2000
plt.subplot(2, 2, 3)
sns.barplot(x=genre_sales_before.index, y=genre_sales_before.values, palette="Blues_r")
plt.title("إجمالي المبيعات حسب النوع قبل 2000")
plt.xticks(rotation=45)
plt.ylabel("مبيعات بالملايين")

# المبيعات بعد 2000
plt.subplot(2, 2, 4)
sns.barplot(x=genre_sales_after.index, y=genre_sales_after.values, palette="Greens_r")
plt.title("إجمالي المبيعات حسب النوع بعد 2000")
plt.xticks(rotation=45)
plt.ylabel("مبيعات بالملايين")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and read the data
df = pd.read_csv("/content/drive/MyDrive/🟢TheFounder/vgsale_1.csv")

# 2. Clean the data
df = df.dropna(subset=['Year', 'Genre', 'Global_Sales'])  # Remove rows with missing Year, Genre, or Global Sales
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')  # Convert Year to numeric, coercing errors to NaN
df = df.dropna(subset=['Year'])  # Remove rows with non-convertible Year values
df['Year'] = df['Year'].astype(int)  # Convert Year to integer type

# 3. Split the data into two periods
df_before_2000 = df[df['Year'] < 2000]  # Data before the year 2000
df_after_2000 = df[df['Year'] >= 2000]  # Data from the year 2000 and onwards

# 4. Analyze the number of games published by genre
genre_count_before = df_before_2000['Genre'].value_counts()  # Count of games by genre before 2000
genre_count_after = df_after_2000['Genre'].value_counts()  # Count of games by genre after 2000

# 5. Analyze total sales by genre
genre_sales_before = df_before_2000.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)  # Total sales by genre before 2000
genre_sales_after = df_after_2000.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)  # Total sales by genre after 2000

# print("📊 Number of games by genre before 2000:")
# print(genre_count_before)
# print("\n📊 Number of games by genre after 2000:")
# print(genre_count_after)

# print("\n💰 Total sales by genre before 2000 (in millions):")
# print(genre_sales_before)
# print("\n💰 Total sales by genre after 2000 (in millions):")
# print(genre_sales_after)

#----------------------
### Visualization ###
# 6. Plot the results
plt.figure(figsize=(15, 6))

# Number of games before 2000
plt.subplot(2, 2, 1)
sns.barplot(x=genre_count_before.index, y=genre_count_before.values, palette="Blues_d")
plt.title("Number of Games Published by Genre Before 2000")
plt.xticks(rotation=45)
plt.ylabel("Number of Games")

# Number of games after 2000
plt.subplot(2, 2, 2)
sns.barplot(x=genre_count_after.index, y=genre_count_after.values, palette="Greens_d")
plt.title("Number of Games Published by Genre After 2000")
plt.xticks(rotation=45)
plt.ylabel("Number of Games")

# Sales before 2000
plt.subplot(2, 2, 3)
sns.barplot(x=genre_sales_before.index, y=genre_sales_before.values, palette="Blues_r")
plt.title("Total Sales by Genre Before 2000")
plt.xticks(rotation=45)
plt.ylabel("Sales in Millions")

# Sales after 2000
plt.subplot(2, 2, 4)
sns.barplot(x=genre_sales_after.index, y=genre_sales_after.values, palette="Greens_r")
plt.title("Total Sales by Genre After 2000")
plt.xticks(rotation=45)
plt.ylabel("Sales in Millions")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plots
```

```python
df_clean = df.dropna(subset=['Genre', 'Global_Sales'])
```

```python
genre_count = df_clean['Genre'].value_counts().reset_index()
genre_count.columns = ['Genre', 'Game_Count']
genre_count
```

```python
genre_sales = df_clean.groupby('Genre')['Global_Sales'].sum().reset_index()
genre_sales = genre_sales.sort_values(by='Global_Sales', ascending=False)
genre_sales
```
