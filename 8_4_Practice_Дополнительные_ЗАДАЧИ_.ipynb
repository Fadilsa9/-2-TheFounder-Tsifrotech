# حدد ثلاثة ناشرين قاموا بنشر أكبر عدد من الألعاب. أنشئ رسمًا بياني عموديًا يُظهر عدد الألعاب المنشورة لكل ناشر على كل منصة (يمكنك استخدام رسم بياني مكدس).

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. تحميل البيانات من ملف CSV
df = pd.read_csv('/content/drive/MyDrive/🟢TheFounder/vgsale_1.csv')

# 2. حذف الصفوف التي تحتوي على قيم مفقودة في عمود الناشر أو المنصة
df_clean = df.dropna(subset=['Publisher', 'Platform'])

# 3. استخراج أسماء أكثر 3 ناشرين قاموا بنشر أكبر عدد من الألعاب
top_publishers = df_clean['Publisher'].value_counts().head(3).index.tolist()

# 4. تصفية البيانات لتشمل فقط الألعاب التي نشرتها هذه الشركات الثلاثة
df_top = df_clean[df_clean['Publisher'].isin(top_publishers)]

# 5. إنشاء جدول محوري:
# الصفوف = الناشرين، الأعمدة = المنصات، القيم = عدد الألعاب
pivot_table = pd.pivot_table(
    df_top,
    index='Publisher',
    columns='Platform',
    values='Name',
    aggfunc='count',
    fill_value=0  # أي منصة لم تنشر عليها ألعاب تُحسب كـ 0
)

# 6. رسم الرسم البياني المكدس (Stacked Bar Chart)
pivot_table.T.plot(
    kind='bar',           # نوع الرسم: أعمدة
    stacked=True,         # لجعل الرسم مكدسًا
    figsize=(14, 7),      # حجم الشكل
    colormap='tab20'      # ألوان مختلفة لكل ناشر
)

# 7. تخصيص العناوين والمحاور
plt.title('أكثر 3 ناشرين - عدد الألعاب لكل منصة (رسم مكدس)')
plt.xlabel('المنصة')
plt.ylabel('عدد الألعاب')
plt.legend(title='الناشر')  # عنوان صندوق التوضيح
plt.xticks(rotation=45)      # تدوير أسماء المنصات لسهولة القراءة
plt.tight_layout()           # تحسين توزيع العناصر داخل الشكل

# 8. عرض الرسم البياني
plt.show()
```

# أنشئ رسومًا بيانية دائرية تُظهر حصص المبيعات الإجمالية في أمريكا الشمالية وأوروبا واليابان (باستخدام عمود "Other Sales") من إجمالي المبيعات العالمية، من 1980 إلى 2000 و من 2000 إلى 2020.


```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. تحميل البيانات
df = pd.read_csv("/content/drive/MyDrive/🟢TheFounder/vgsale_1.csv")

# 2. حذف الصفوف التي لا تحتوي على سنة إصدار
df = df.dropna(subset=['Year'])

# 3. تحويل عمود السنة إلى أرقام صحيحة (لأنها قد تكون بصيغة float)
df['Year'] = df['Year'].astype(int)

# 4. تقسيم البيانات إلى فترتين: قبل 2000 وبعد 2000
df_1980_2000 = df[(df['Year'] >= 1980) & (df['Year'] < 2000)]
df_2000_2020 = df[(df['Year'] >= 2000) & (df['Year'] <= 2020)]

# 5. دالة لحساب مجموع المبيعات لكل منطقة في فترة معينة
def sales_share_by_region(df_period):
    na = df_period['NA_Sales'].sum()
    eu = df_period['EU_Sales'].sum()
    jp = df_period['JP_Sales'].sum()
    other = df_period['Other_Sales'].sum()
    return [na, eu, jp, other]

# 6. حساب الحصص لكل فترة
shares_1980_2000 = sales_share_by_region(df_1980_2000)
shares_2000_2020 = sales_share_by_region(df_2000_2020)

# 7. إعداد أسماء المناطق
labels = ['North America', 'Europe', 'Japan', 'Other']

# 8. إنشاء رسمين دائريين
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# 9. رسم الحصة من 1980 إلى 2000
axs[0].pie(shares_1980_2000, labels=labels, autopct='%1.1f%%', startangle=140)
axs[0].set_title('Sales Share (1980–2000)')

# 10. رسم الحصة من 2000 إلى 2020
axs[1].pie(shares_2000_2020, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1].set_title('Sales Share (2000–2020)')

# 11. تحسين التخطيط العام
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/🟢TheFounder/IQ_countries.csv")
df.head()
```

```python
print(df.info())
```

```python
df.describe()
```

```python
# تحويل الأعمده الرقمية إلى numeric

columns_to_fix = ['Average IQ', 'Literacy Rate', 'Nobel Prices',
                  'Human Development Index', 'Mean years of schooling',
                  'Gross National Income', 'Population']

for col in columns_to_fix:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.info()
```

```python
# معالجة القيم الفقودة

print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.head()
```

```python
sns.histplot(df['Average IQ'], kde=True, color='skyblue')
plt.title('Distribution of Average IQ')
plt.xlabel('Average IQ')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['Average IQ'], color='skyblue')
```

```python
sns.barplot(x='Continent', y='Average IQ', data=df)
plt.title("Average IQ by Continent")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x='Continent', y='Average IQ', data=df)
plt.title('Average IQ by Continent')
plt.xlabel('Continent')
plt.ylabel('Average IQ')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
