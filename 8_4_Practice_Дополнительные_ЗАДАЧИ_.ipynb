```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and read the data
df = pd.read_csv("/content/drive/MyDrive/üü¢TheFounder/vgsale_1.csv")

# 2. Clean the data
df = df.dropna(subset=['Year', 'Genre', 'Global_Sales'])  # Remove rows with missing Year, Genre, or Global Sales
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')  # Convert Year to numeric, coercing errors to NaN
df = df.dropna(subset=['Year'])  # Remove rows with non-convertible Year values
df['Year'] = df['Year'].astype(int)  # Convert Year to integer type

# 3. Split the data into two periods
df_before_2000 = df[df['Year'] < 2000]  # Data before the year 2000
df_after_2000 = df[df['Year'] >= 2000]  # Data from the year 2000 and onwards

# 4. Analyze the number of games published by genre
genre_count_before = df_before_2000['Genre'].value_counts()  # Count of games by genre before 2000
genre_count_after = df_after_2000['Genre'].value_counts()  # Count of games by genre after 2000

# 5. Analyze total sales by genre
genre_sales_before = df_before_2000.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)  # Total sales by genre before 2000
genre_sales_after = df_after_2000.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)  # Total sales by genre after 2000

# print("üìä Number of games by genre before 2000:")
# print(genre_count_before)
# print("\nüìä Number of games by genre after 2000:")
# print(genre_count_after)

# print("\nüí∞ Total sales by genre before 2000 (in millions):")
# print(genre_sales_before)
# print("\nüí∞ Total sales by genre after 2000 (in millions):")
# print(genre_sales_after)

#----------------------
### Visualization ###
# 6. Plot the results
plt.figure(figsize=(15, 6))

# Number of games before 2000
plt.subplot(2, 2, 1)
sns.barplot(x=genre_count_before.index, y=genre_count_before.values, palette="Blues_d")
plt.title("Number of Games Published by Genre Before 2000")
plt.xticks(rotation=45)
plt.ylabel("Number of Games")

# Number of games after 2000
plt.subplot(2, 2, 2)
sns.barplot(x=genre_count_after.index, y=genre_count_after.values, palette="Greens_d")
plt.title("Number of Games Published by Genre After 2000")
plt.xticks(rotation=45)
plt.ylabel("Number of Games")

# Sales before 2000
plt.subplot(2, 2, 3)
sns.barplot(x=genre_sales_before.index, y=genre_sales_before.values, palette="Blues_r")
plt.title("Total Sales by Genre Before 2000")
plt.xticks(rotation=45)
plt.ylabel("Sales in Millions")

# Sales after 2000
plt.subplot(2, 2, 4)
sns.barplot(x=genre_sales_after.index, y=genre_sales_after.values, palette="Greens_r")
plt.title("Total Sales by Genre After 2000")
plt.xticks(rotation=45)
plt.ylabel("Sales in Millions")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plots
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the dataset from the CSV file
df = pd.read_csv('/content/drive/MyDrive/üü¢TheFounder/vgsale_1.csv')

# 2. Drop rows where 'Publisher' or 'Platform' is missing (NaN)
df_clean = df.dropna(subset=['Publisher', 'Platform'])

# 3. Identify the top 3 publishers with the most games published
top_publishers = df_clean['Publisher'].value_counts().head(3).index.tolist()

# 4. Filter the dataset to include only games by the top 3 publishers
df_top = df_clean[df_clean['Publisher'].isin(top_publishers)]

# 5. Create a pivot table:
# Rows = Publisher, Columns = Platform, Values = Count of games (by 'Name')
pivot_table = pd.pivot_table(
    df_top,
    index='Publisher',
    columns='Platform',
    values='Name',
    aggfunc='count',
    fill_value=0  # Platforms with no games will be filled with zero
)

# 6. Plot the data using a stacked bar chart
pivot_table.T.plot(
    kind='bar',           # Bar chart
    stacked=True,         # Enable stacked bars
    figsize=(14, 7),      # Set figure size
    colormap='tab20'      # Use a colorful colormap for distinction
)

# 7. Customize the chart labels and appearance
plt.title('Top 3 Publishers - Number of Games per Platform (Stacked)')
plt.xlabel('Platform')
plt.ylabel('Number of Games')
plt.legend(title='Publisher')       # Add legend with title
plt.xticks(rotation=45)             # Rotate x-axis labels for readability
plt.tight_layout()                  # Automatically adjust layout

# 8. Show the plot
plt.show()
```

# ÿ£ŸÜÿ¥ÿ¶ ÿ±ÿ≥ŸàŸÖŸãÿß ÿ®ŸäÿßŸÜŸäÿ© ÿØÿßÿ¶ÿ±Ÿäÿ© ÿ™Ÿèÿ∏Ÿáÿ± ÿ≠ÿµÿµ ÿßŸÑŸÖÿ®Ÿäÿπÿßÿ™ ÿßŸÑÿ•ÿ¨ŸÖÿßŸÑŸäÿ© ŸÅŸä ÿ£ŸÖÿ±ŸäŸÉÿß ÿßŸÑÿ¥ŸÖÿßŸÑŸäÿ© Ÿàÿ£Ÿàÿ±Ÿàÿ®ÿß ŸàÿßŸÑŸäÿßÿ®ÿßŸÜ (ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿπŸÖŸàÿØ "Other Sales") ŸÖŸÜ ÿ•ÿ¨ŸÖÿßŸÑŸä ÿßŸÑŸÖÿ®Ÿäÿπÿßÿ™ ÿßŸÑÿπÿßŸÑŸÖŸäÿ©ÿå ŸÖŸÜ 1980 ÿ•ŸÑŸâ 2000 Ÿà ŸÖŸÜ 2000 ÿ•ŸÑŸâ 2020.


```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
df = pd.read_csv("/content/drive/MyDrive/üü¢TheFounder/vgsale_1.csv")

# 2. ÿ≠ÿ∞ŸÅ ÿßŸÑÿµŸÅŸàŸÅ ÿßŸÑÿ™Ÿä ŸÑÿß ÿ™ÿ≠ÿ™ŸàŸä ÿπŸÑŸâ ÿ≥ŸÜÿ© ÿ•ÿµÿØÿßÿ±
df = df.dropna(subset=['Year'])

# 3. ÿ™ÿ≠ŸàŸäŸÑ ÿπŸÖŸàÿØ ÿßŸÑÿ≥ŸÜÿ© ÿ•ŸÑŸâ ÿ£ÿ±ŸÇÿßŸÖ ÿµÿ≠Ÿäÿ≠ÿ© (ŸÑÿ£ŸÜŸáÿß ŸÇÿØ ÿ™ŸÉŸàŸÜ ÿ®ÿµŸäÿ∫ÿ© float)
df['Year'] = df['Year'].astype(int)

# 4. ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ ŸÅÿ™ÿ±ÿ™ŸäŸÜ: ŸÇÿ®ŸÑ 2000 Ÿàÿ®ÿπÿØ 2000
df_1980_2000 = df[(df['Year'] >= 1980) & (df['Year'] < 2000)]
df_2000_2020 = df[(df['Year'] >= 2000) & (df['Year'] <= 2020)]

# 5. ÿØÿßŸÑÿ© ŸÑÿ≠ÿ≥ÿßÿ® ŸÖÿ¨ŸÖŸàÿπ ÿßŸÑŸÖÿ®Ÿäÿπÿßÿ™ ŸÑŸÉŸÑ ŸÖŸÜÿ∑ŸÇÿ© ŸÅŸä ŸÅÿ™ÿ±ÿ© ŸÖÿπŸäŸÜÿ©
def sales_share_by_region(df_period):
    na = df_period['NA_Sales'].sum()
    eu = df_period['EU_Sales'].sum()
    jp = df_period['JP_Sales'].sum()
    other = df_period['Other_Sales'].sum()
    return [na, eu, jp, other]

# 6. ÿ≠ÿ≥ÿßÿ® ÿßŸÑÿ≠ÿµÿµ ŸÑŸÉŸÑ ŸÅÿ™ÿ±ÿ©
shares_1980_2000 = sales_share_by_region(df_1980_2000)
shares_2000_2020 = sales_share_by_region(df_2000_2020)

# 7. ÿ•ÿπÿØÿßÿØ ÿ£ÿ≥ŸÖÿßÿ° ÿßŸÑŸÖŸÜÿßÿ∑ŸÇ
labels = ['North America', 'Europe', 'Japan', 'Other']

# 8. ÿ•ŸÜÿ¥ÿßÿ° ÿ±ÿ≥ŸÖŸäŸÜ ÿØÿßÿ¶ÿ±ŸäŸäŸÜ
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# 9. ÿ±ÿ≥ŸÖ ÿßŸÑÿ≠ÿµÿ© ŸÖŸÜ 1980 ÿ•ŸÑŸâ 2000
axs[0].pie(shares_1980_2000, labels=labels, autopct='%1.1f%%', startangle=140)
axs[0].set_title('Sales Share (1980‚Äì2000)')

# 10. ÿ±ÿ≥ŸÖ ÿßŸÑÿ≠ÿµÿ© ŸÖŸÜ 2000 ÿ•ŸÑŸâ 2020
axs[1].pie(shares_2000_2020, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1].set_title('Sales Share (2000‚Äì2020)')

# 11. ÿ™ÿ≠ÿ≥ŸäŸÜ ÿßŸÑÿ™ÿÆÿ∑Ÿäÿ∑ ÿßŸÑÿπÿßŸÖ
plt.tight_layout()
plt.show()
```
